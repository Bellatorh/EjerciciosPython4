```python
import random

#Esta línea define una función llamada simulador_clima, que es el corazón del programa que 
#va a contener los parametros que el usuario ingrese
def simulador_clima(temperatura_inicial, probabilidad_lluvia, dias):
    # Variables iniciales
    temperatura = temperatura_inicial
    prob_lluvia = probabilidad_lluvia
    #Crea una lista vacía donde se registrarán las temperaturas de cada día = []
    temperaturas = []
    # Lleva un contador de los días en los que llueve durante la simulación = 0
    dias_lluvia = 0

    # Simulación diaria
    '''La función range(inicio, fin) genera una secuencia de números desde inicio hasta uno antes de fin (el valor final no se incluye)
    Cuando indicamos dias + 1 como el segundo argumento, aseguramos que el bucle se ejecute para el número exacto de días que pidió el 
    usuario. Sin el +1, el último día se quedaría fuera Al agregar +1 como en range(1, dias + 1), generamos: 1, 2, 3, 4, 5, asegurándonos
    de incluir todos los días'''
    for dia in range(1, dias + 1):
        print(f"\nDía {dia}:")
        
        # Variación de temperatura
        '''Selecciona aleatoriamente entre los valores -2 y 2. Esto representa la variación diaria de la temperatura (sube o baja 2 grados)'''
        cambio_temp = random.choice([-2, 2])
        '''Modifica la temperatura actual sumando el valor de cambio_temp'''
        temperatura += cambio_temp
        print(f"La temperatura cambia en {cambio_temp}°C y ahora es {temperatura}°C.")

        # Ajustes en probabilidad de lluvia
        '''hay un límite lógico en la probabilidad de lluvia, que debe estar entre 0% y 100%. entonces aquí el uso de min y max
        para garantizar que la probabilidad no pase de 100% o que no sea menor de 0%'''
        if temperatura > 25:
            prob_lluvia = min(100, prob_lluvia + 20)  # Máximo 100%
            print(f"La temperatura superó los 25°C. La probabilidad de lluvia aumenta al {prob_lluvia}%.")
        elif temperatura < 5:
            prob_lluvia = max(0, prob_lluvia - 20)  # Mínimo 0%
            print(f"La temperatura bajó de 5°C. La probabilidad de lluvia disminuye al {prob_lluvia}%.")

        # Determinar si llueve
        llueve = random.randint(1, 100) <= prob_lluvia
        if llueve:
            dias_lluvia += 1
            temperatura -= 1  # Baja 1°C si llueve
            print(f"¡Llueve! La temperatura baja 1°C y ahora es {temperatura}°C.")
        else:
            print("No llueve hoy.")

        # Guardar temperatura diaria
        temperaturas.append(temperatura)
    
    # Resultados
    temp_max = max(temperaturas)
    temp_min = min(temperaturas)

    print("\nResultados de la simulación:")
    print(f"Temperatura máxima: {temp_max}°C")
    print(f"Temperatura mínima: {temp_min}°C")
    print(f"Días con lluvia: {dias_lluvia}")
    print("Temperaturas diarias:", temperaturas)

# Pedir datos al usuario
print("Bienvenido al simulador de clima.")
temperatura_inicial = int(input("Ingrese la temperatura inicial (en °C): "))
probabilidad_lluvia = int(input("Ingrese la probabilidad inicial de lluvia (en %): "))
dias = int(input("Ingrese el número de días a simular: "))

# Llamar a la función con los datos ingresados
simulador_clima(temperatura_inicial, probabilidad_lluvia, dias)
```
